```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# Import data
df = pd.read_csv("https://uit-sok-1006-v23.github.io/data/suits_watermelons.csv", delimiter=";")

# Create variables using iloc function in the numpy package and target the right column
df['p'] = np.exp(df.iloc[:,3])
df['h'] = np.exp(df.iloc[:,2])
df['q'] = np.exp(df.iloc[:,1])

# Plot for prices of watermelons
plt.plot(df.iloc[:,0], df['p'])
plt.title('Price of watermelons')
plt.xlabel('Year')
plt.ylabel('Price')
plt.show()

# Plot for harvest and crop quantities 
plt.plot(df.iloc[:,0], df['q'], 'o--', label='Crop')
plt.plot(df.iloc[:,0], df['h'], 'o-', label='Harvest')
plt.title('Harvest and crop quantities')
plt.xlabel('Year')
plt.ylabel('Quantity')
plt.legend()
plt.show()
```

```python
#Define the variables needed for the table
Q = np.arange(20, 101, 5)
log_Q = np.log(Q)
Demand = 8.5 - (0.82 * log_Q)
Supply = -2.0 + (1.7 * log_Q)
Supply_price = np.exp(Supply)
Demand_price = np.exp(Demand)

new_supply = Supply + 0.4
log_new_supply = np.exp(new_supply)

#Create a table with the data
table = pd.DataFrame({'Quantity': Q,
                      'Log_Quantity': log_Q,
                      'Log_Demand': Demand,
                      'Log_Supply': Supply,
                      'Supply_Price': Supply_price,
                      'Demand_price': Demand_price,
                      'New_Supply': new_supply,
                      'Log_New_Supply': log_new_supply})

#Print the table
table
```

```python
#Plot the supply and demand curves
plt.plot(Q, Supply_price, 'o-', label='Supply')
plt.plot(Q, Demand_price, 'o-', label='Demand')

#make the plot look nice
plt.legend()
plt.title('Supply and Demand curve')
plt.xlabel('Quantity in millions')
plt.ylabel('Price per thousands')
```

```python
#make the same plot as above, but with the new supply curve
plt.plot(Q, Supply_price, 'o-', label='Supply')
plt.plot(Q, Demand_price, 'o-', label='Demand')

#new supply curve
plt.plot(Q, log_new_supply, 'o-', label='New Supply')

#make the plot look nice
plt.legend()
plt.title('Supply and Demand curve')
plt.xlabel('Quantity in millions')
plt.ylabel('Price per thousands')
```

```python
#calculate the elasticity of supply and demand

Eq_s = sp.Eq(-2.0 + 1.7*q, p)
Eq_d = sp.Eq(8.5 - 0.82*q, p)
solve_s = sp.solve(Eq_s, q)
solve_d = sp.solve(Eq_d, q)

#display the results
display(solve_s, solve_d)

#print the results and explain
print(f'''
    Priselastisiteten til etterspørselen er -1.219. 
    Dette vil si at om vi øker prisen med 1% så reduseres etterspørselen med circa 1.22%.
    Etterspørselen er elastisk. som vil si at den er følsom for pris endringer.
      
    Priselastisiteten til tilbudet er 0.588. Dersom vi øker prisen med 1%, økes tilbudet med 0.588%. 
    Tilbudet er uelastisk som betyr at det er  følsomt for prisendringer.
      ''')
```
